## 7. Sorting Data

```python
people = {
    'First':['Asim','Vijay','Saud','Muzammil','Sagar'],
    'Last':['Fakhi','Yadav','Fakhi','Daire','Kothari'],
    'Email':['Asim.F@company.com','Vijay.Y@company.com','Saud.F@company.com','Muzammil.D@company.com','Sagar.K@company.com']
}
```

```python
df = pd.DataFrame(people)
df
```

```python
# We can sort the above dataframe by any column we want can be ascending or descending.
# In order to achieve this we can use sort_value() method and passs the column name as a 'by' parameter
df.sort_values(by='Last',ascending=False) #Setting Ascending to Flase will make it Descending
```

```python
# Now if I want to do sort by last name first folowed by sorting of first name if it has an identical last name.
df.sort_values(by=['Last','First'],ascending=True)
```

```python
# Now if I wan to sort one column asce but other one as desc.
#To achieve this we can use a dictionary of bools in order of the columns sorting type
df.sort_values(by=['Last','First'],ascending=[False,True],inplace=True)
df
```

```python
# Now that our DataFrame is modified and if we want to reset the sorting by index column we can do it using sort_index()
df.sort_index()
```

```python
#It is also possible to sort a particulart column only
df['First'].sort_values()
```

## 8.Grouping and Aggregating - Analyzing and Exploring Data

When you think of analyzing the data the main thing that anylys do is grouping  and aggregation the data.

```python
# To see the median of the salaries that a developer is getting in this survey we can use .median() function
df['ConvertedComp'].median()
```

```python
#Or if we use median to the complete dataset then it will give us median of only numerical columns
df.median()
```

```python
# If we want to get broad overview of our data then we ca simply use describe() fntion
df.describe()
```

Its is preferred to use meadian as mean could be affected by outliers very heavil. And we have observed this in case of our .nlargest() and .nsmallest().When we say numeric columns the its going to skip the nan values too.
