# Line Plots

```python
linear_data = np.array([1,2,3.4,5,6,7,8])
quadratic_data = linear_data**2
plt.figure()
plt.plot(linear_data,'-o',quadratic_data,'-o')
plt.plot([22,44,55],'--r')
plt.xlabel('Some Data')
plt.ylabel('Some other Data')
plt.title('An Awesome Line Plot')
plt.legend(['Baseline','Competition','Us'],loc=2)
plt.gca().fill_between(range(len(linear_data)),
                       linear_data, quadratic_data,
                       facecolor='blue',
                       alpha=0.25)
```

Here we can only give y axis values to our plot call  but no x axis values.<br>
Instead the plot fucntion decided the scale itself.<br>
also the color of the lines for different data points.<br>

```python
import pandas as pd
plt.figure()
observation_dates = np.arange('2017-01-01','2017-01-08',dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime, observation_dates))
plt.plot(observation_dates, linear_data,'-o',
        observation_dates, quadratic_data,'-o')
```

```python
plt.figure()
plt.style.use('fivethirtyeight')
observation_dates = np.arange('2017-01-01','2017-01-08',dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime, observation_dates))
plt.plot(observation_dates, linear_data,'-o',
        observation_dates, quadratic_data,'-o')
x = plt.gca().xaxis # get the x axis and then rotate each ticklabel.
for item in x.get_ticklabels():
    item.set_rotation(45)
plt.subplots_adjust(bottom=.25)
ax = plt.gca()
ax.set_xlabel('Dates')
ax.set_ylabel('Units')
ax.set_title('Quadratic vs Linear Performance')
```

# Bar Charts

```python
plt.figure()
xvals = range(len(linear_data))
plt.bar(xvals, linear_data, width=0.3)
```

```python
plt.figure()
xvals = range(len(linear_data))
plt.bar(xvals, linear_data, width=0.3)
new_xvals = []
for item in xvals:
    new_xvals.append(item+0.3)
plt.bar(new_xvals,quadratic_data,width=0.3,color='red')
```

```python
# We simply add error value inside the bar grah itself
from random import randint
linear_err = [randint(0,10) for x in range(len(linear_data))]
plt.bar(xvals, linear_data, width=0.3,yerr=linear_err)
```

```python
# We can also creat stacked bar chart by adding second set of data on top of first
plt.figure()
xvals = range(len(linear_data))
plt.bar(xvals,linear_data, width=0.3,color='b')
plt.bar(xvals,quadratic_data, width=0.3, bottom=linear_data,color='y')
```

```python
# We can also pivot this bar graph to a horizontal bar graph by calling the barh function
plt.figure()
xvals = range(len(linear_data))
plt.barh(xvals, linear_data,height=0.3,color='b')
plt.barh(xvals,quadratic_data, height=0.3, left=linear_data,color='y')
```

# Dejunkifying a plot

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

# change the bar colors to be less bright blue
bars = plt.bar(pos, popularity, align='center', linewidth=0, color='lightslategrey')
# make one bar, the python bar, a contrasting color
bars[0].set_color('#1F77B4')

# soften all labels by turning grey
plt.xticks(pos, languages, alpha=0.8)
plt.ylabel('% Popularity', alpha=0.8)
plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=0.8)

# remove all the ticks (both axes), and tick labels on the Y axis
plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')

# remove the frame of the chart
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.show()
```
